```python
# INCREMENT PERCENTAGE
co2em.drop(['Unnamed: 61'], axis = 1, inplace = True)
```

```python
co2em.head()
```

```python
co2em['1960' : '2013'].head()
```

```python
# use loc for slicing operator#
# : sselects al rows
# '1960' : '2013' slices that colums
co2em = co2em.loc[:, '1960' : '2013']
```

```python
co2em.head()
```

```python
# give key value for dictionary

co2em['Percentage'] = 90
```

```python
co2em.head()
```

```python
# to determine the growth 
# - to check the change in the value
# to check the change in data = (final - intial) / intial 
# kati times la growth vhai rako ho vhanera nikalna khojeko 
# np.round , 2=> rounds the data in 2 decimal 

co2em['Percentage'] = np.round((co2em['2013'] - co2em['1960']) / co2em['1960'], 2)
```

```python
co2em.head()
```

```python
change = co2em['Percentage']
```

```python
change.head()
```

```python
change = change.dropna().sort_values(ascending = False) # this sorts in descending order
```

```python
# increaseing co2 emmission level
change.head()
```

```python
# decreasing co2 emmission level
change.tail()
```

```python
change.hist()
```

```python
change.plot(kind = 'bar', figsize = (60, 26))
```

```python
# is an outlier with max value [' United Arab Emirates']
# skips first 10 items with high co2 emission 
change.iloc[10:].plot(kind = 'bar', figsize = (60, 26))
```
