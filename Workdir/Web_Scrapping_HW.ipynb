```python
for rows in maindiv.find('table').find('tbody').find_all('tr'):
    if rows.find('th'):
        continue
    tds = rows.find_all('td')
    print('Airlines', tds[0].text)
    print('STA', tds[1].text)
    print('ETA', tds[2].text)
    print('Flight', tds[3].text)
    print('Origin', tds[4].text)
    print('Status', tds[5].text)
    break
```

```python
import csv
```

```python
with open('web.txt', 'w') as webfile:
    csv_writer = csv.DictWriter(webfile, fieldnames = ['Airlines', 'STA', 'ETA', 'Flight', 'Origin', 'Status'])
    csv_writer.writeheader()
    
    data = []
    
    for rows in maindiv.find('table').find('tbody').find_all('tr'):
        _rowdata = {}
        if rows.find('th'):
            continue
        tds = rows.find_all('td')
        
        _rowdata['Airlines'] = tds[0].text.strip()
        _rowdata['STA'] = tds[1].text.strip()
        _rowdata['ETA'] = tds[2].text.strip()
        _rowdata['Flight'] = tds[3].text.strip()
        _rowdata['Origin'] = tds[4].text.strip()
        _rowdata['Status'] = tds[5].text.strip()
        
        data.append(_rowdata)
        
    csv_writer.writerows(data)
```
